Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (12, 9)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
emission_df['VEHICLE CLASS'].describe()
```

# **Exploring the relationship between CO2 Emissions and Engine Size of vehicles**

```python
plt.figure(figsize=(18, 6))

sns.scatterplot(x= emission_df.ENGINE_SIZE, y= emission_df.CO2_EMISSIONS, hue =emission_df['VEHICLE CLASS']);
plt.title('Engine Size vs CO2 Emissions');
```

The above scatter plot shows the relationship between CO2 Emissions and Engine Size of vehicles for different vehicle classes

# **Exploring the dataset of top 5 vehicles that cause the CO2 emissions**

```python
maxemission_df
```

```python
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (17, 15)
matplotlib.rcParams['figure.facecolor'] = '#00000000'

sns.barplot(x= 'MAKE', y = 'CO2_EMISSIONS' , hue ='MODEL.1', data=maxemission_df);
plt.title('Top 5 vehicles causing CO2 emissions');
```

From the above Bar graph it is clear that the **GMC K1500 SIERRA 4X4** has the maximum CO2 emission

# **Exploring the fuel consumption distribution of VOLVO cars using Histogram**

```python
emission_df
```

```python
volvo_df = emission_df[emission_df.MAKE == 'VOLVO']
volvo_df
```

```python
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (10, 7)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
sns.histplot(volvo_df['FUEL_CONSUMPTION*'] , bins =5);
plt.title('fuel consumption distribution of VOLVO cars');
```

From the above Histogram it is observed that there are around **11** cars having Fuel Consumption of **12.5 to 13.1**
